### Show both text and visual confusion Matrices using scikit learn and matplotlib and explain what the graph tells you and what you did. (2.5) 


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_regression_results(y_true, y_pred, title='Regression Results'):

    fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(7, 10))
    
    # Plotting actual vs. predicted values
    sns.scatterplot(x=y_true, y=y_pred, ax=ax[0])
    ax[0].plot([min(y_true), max(y_true)], [min(y_true), max(y_true)], '--', color='red')  
    ax[0].set_title('Actual vs. Predicted Values')
    ax[0].set_xlabel('Actual Values')
    ax[0].set_ylabel('Predicted Values')

    # Plotting residuals
    residuals = y_true - y_pred
    sns.scatterplot(x=y_pred, y=residuals, ax=ax[1])
    ax[1].axhline(y=0, linestyle='--', color='red')  
    ax[1].set_title('Residuals')
    ax[1].set_xlabel('Predicted Values')
    ax[1].set_ylabel('Residuals')

    plt.tight_layout()
    plt.suptitle(title, y=1.05)
    plt.show()


#reshape test to ensure it's 1 dimensional 
BSHy_test = BSHy_test.ravel()

#plot the regression results
plot_regression_results(y_true=BSHy_test, y_pred=BSHy_pred)


```

The graphs show that the gradient boosting regression works fairly well to predict the counts for bikeshare. Although the residuals became larger at the middle range of the predicted values.
